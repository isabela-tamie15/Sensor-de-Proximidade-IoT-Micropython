# **Sensor de Proximidade com LEDs Progressivos e Acionamento de Relé**
---
## **Descrição do Projeto**

Este projeto implementa um **sistema de detecção de proximidade utilizando sensor ultrassônico**, com resposta visual progressiva por LEDs e acionamento de um relé quando o objeto atinge uma distância crítica.

O sistema foi desenvolvido em **MicroPython para ESP32** e simulado na plataforma Wokwi, demonstrando conceitos fundamentais de **sistemas embarcados, sensores e atuadores**.

A lógica implementa três níveis de proximidade e simula aplicações reais como:

* sensores de estacionamento
* sistemas de segurança
* automação industrial
* robótica móvel

## **Funcionamento do Sistema**

O sensor ultrassônico mede a distância de um objeto utilizando o **tempo de retorno de um pulso sonoro**.

O microcontrolador calcula a distância e ativa indicadores visuais progressivos.

### Níveis de proximidade

| Distância | Comportamento       |
| --------- | ------------------- |
| < 70 cm   | LED verde acende    |
| < 40 cm   | LED amarelo acende  |
| < 20 cm   | LED vermelho acende e  Relé é acionado|

O relé representa o acionamento de um dispositivo externo, como um motor, alarme ou sistema automatizado.

---

## **Arquitetura do Sistema**

O fluxo do sistema ocorre da seguinte forma:

1. O ESP32 envia um pulso pelo pino **Trigger** do sensor ultrassônico
2. O sensor emite uma onda sonora
3. O pino **Echo** recebe o tempo de retorno da onda
4. O sistema calcula a distância com base na velocidade do som
5. A lógica condicional define quais LEDs devem ser acionados
6. Se a distância for crítica, o relé é ativado

## **Lógica de Cálculo da Distância**

A distância é calculada a partir do tempo de retorno do pulso ultrassônico utilizando a fórmula:

Distância = (Tempo × Velocidade do Som) / 2

Onde:

* velocidade do som ≈ **0.0343 cm/µs**
* divisão por 2 ocorre devido ao trajeto **ida e volta da onda sonora**

No código, isso é implementado da seguinte forma:

```python
distancia = (duracao * 0.0343) / 2
```
